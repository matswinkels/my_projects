## 2. Prezentacja podstawowych możliwości pakietu

### Tworzenie podstawowego wykresu

Wykresy tworzone przy użyciu pakietu matplotlib są umieszczane w obiektach typu **figure**. Są to niejako "pojemniki" na pozostałe elementy wykresu.

Tworzenie wykresu odbywa się za pomocą funkcji **figure()**. Z kolei do wyświetlenia wykresu służy funkcja **show()**.

Do rysowania podstawowego wykresu, poprzez podanie danych do naniesienia, używa się funkcji **plot()**. W poniższym przykładzie poddałem wizualizacji funkcję sinusa.

```python
x = np.arange(0, np.pi *2, 0.05)
y = np.sin(x)
```

```python
fig1 = plt.figure(figsize=(6.4, 4.8),
                 facecolor='lightblue',
                 edgecolor='blue',
                 linewidth=5)
# Parametry funkcji:
# figsize - wielkość wykresu w calach
# dpi - rozdzielczość wykresu
# facecolor, edgecolor - kolor tła, kolor obramowania wykresu
# linewidth - szerokość obramowania

plt.plot(x, y)

plt.show()
```

### Dodawanie i edycja elementów wykresu

```python
fig1 = plt.figure(figsize=(9.6, 7.2),
                 facecolor='#7fbfdb')

# ZMIANA STYLU I SZEROKOŚCI LINII:
plt.plot(x, y, '--', lw=2, color='r')
# EDYCJA ELEMENTÓW WYKRESU:
plt.title('Wykres funkcji sinus', fontsize=20)
plt.xlabel('Kąt', fontsize=16)
plt.ylabel('Sinus', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# DODANIE SIATKI:
plt.grid(linestyle='--', alpha=0.6, linewidth=0.5, fillstyle='left')

plt.show()
```

### WIzualizacja wartości na wspólnym wykresie

Jeżeli chcemy na jednym wykresie zwizualizować więcej serii danych o wspólnej osi, wystarczy użyć funkcji **plot()**, by nanieść dane. Do rozróżnienia danych stosuje się legendę - funkcja **legend()**.

```python
x = np.arange(-2*np.pi, 2*np.pi, 0.05)
y = np.sin(x)
z = np.cos(x)
```

```python
fig2 = plt.figure(figsize=(9.6, 7.2))

plt.plot(x, y, '-', lw=2, color='r')
# NANIESIENIE DRUGIEJ SERII DANYCH:
plt.plot(x, z, '-', lw=2, color='b')

plt.title('Wykres funkcji sinus i cosinus', fontsize=20)
plt.xlabel('Kąt', fontsize=16)
plt.ylabel('Wartość', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# DODANIE POZIOMEJ (dla y=0) I PIONOWEJ (dla x=0) LINII:
plt.axhline(y=0, linestyle='--', color='black', alpha=0.2)
plt.axvline(x=0, linestyle='--', color='black', alpha=0.2)

# DODANIE LEGENDY:
plt.legend(['sin', 'cos'], loc='lower left', fontsize=16, framealpha=0.4)
plt.grid(linestyle='--', alpha=0.6, linewidth=0.5, fillstyle='left')

plt.show()
```

### Klasa osi (Axes Class)

Traktowanie osi wykresu jako obiektu. Jeden wykres (figure) może posiadać wiele osi, jednak konkretna oś może być składową tylko jednego wykresu.

Dodanie osi do wykresu następuję przy użyciu metody **add_subplot(liczba wierszy siatki osi, liczba kolumn siatki osi, indeks)**.

Do każdej z osi można dotrzeć używając metod, np. set_title, set_xlabel, set_ylabel, grid itp.

```python
fig3 = plt.figure(figsize=(13,8))

# STWORZENIE OSI / DODANIE OSI DO WYKRESU:
ax1 = fig3.add_subplot(2, 2, 1)
ax2 = fig3.add_subplot(2, 2, 2)
ax3 = fig3.add_subplot(2, 2, 3)

# DOKONANIE WIZUALIZACJI DANYCH NA POSZCZEGÓLNYCH OSIACH
ax1.hist(np.random.randn(100), bins = 20, color = 'k', alpha = 0.3)
ax2.plot(np.random.randn(50).cumsum(), 'k--')
ax3.scatter(np.arange(30), np. arange(30) + 3 * np.random.randn(30))

ax1.set_title('Histogram', fontsize=16)
ax2.set_title('Line Chart', fontsize=16)
ax3.set_title('Scatter plot', fontsize=16)

plt.show()
```

### Dodawanie wykresów składowych (sublots)

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(13,8))
# Parametry 'sharex' oraz 'sharey' definiują, czy wyrazy składowe mają
# dzielić ze sobą osie X oraz Y.

# STWORZENIE HISTOGRAMU DLA KAŻDEGO WYRAZU SKŁADOWEGO:
for i in range(len(axes)):
    for j in range(len(axes)):
        axes[i, j].hist(np.random.randn(500),
                        bins=50,
                        color='b',
                        alpha=0.5)
# MODYFIKACJA MARGINESÓW WYKRESÓW SKŁADOWYCH:
plt.subplots_adjust(wspace=0, hspace=0)
        
```
